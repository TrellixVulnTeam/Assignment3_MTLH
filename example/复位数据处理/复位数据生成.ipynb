# 处理数据

## 生成路径字典

```python
folder_path = "E:/futures/futures_daily/889"
full_file_dict = {file: folder_path+"/"+file for file in os.listdir(folder_path)}
```

## 导入数据并处理

book_id改为underlying_symbol，并仅保留品种代码，并且从index转为普通的列。\
date改为datetime，并且从index转为普通的列。\
total_turnover改为turnover。\
limit_down改为lower_limit。\
limit_up改为upper_limit。\
dominant_id改为contract

```python
df_list = []
```

```python
for file in tqdm(full_file_dict):
    df = pd.read_pickle(full_file_dict[file])
    df = df.reset_index()
    df = df.rename(columns={'date': 'datetime', 'total_turnover': 'turnover',\
                           'limit_down':'lower_limit','limit_up':'upper_limit','dominant_id':'contract'})
    df_list.append(df)
```

```python
data = pd.concat(df_list, axis=0)
```

```python
data
```
