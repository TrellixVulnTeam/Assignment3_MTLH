# 统计商品池每日进出和每个品种进出的次数

```python
self = CommodityPoolManager()
dynamic_pool3_in_out_dict= self.get_commodity_pool_in_out(group='DynamicPool', name='DynamicPool3')
```

```python
series1 = pd.Series(dynamic_pool3_in_out_dict['symbol_in_num']).sort_values(ascending=False)
series1.to_csv("D:/LFProjects/NewPythonProject/example/商品池/dynamic_pool3_in_num.csv")
```

```python
plt.figure(figsize=(20, 8))
plt.bar(x=series1.index, height=series1.values)
plt.title("DynamicPool3品种进入次数")
plt.show()
```

```python
series2 = pd.Series(dynamic_pool3_in_out_dict['symbol_out_num']).sort_values(ascending=False)
series2.to_csv("D:/LFProjects/NewPythonProject/example/商品池/dynamic_pool3_out_num.csv")
```

```python
plt.figure(figsize=(20, 8))
plt.bar(x=series2.index, height=series2.values)
plt.title("DynamicPool3品种退出次数")
plt.show()
```
