# BaseClass

BaseClass是一个用于管理参数的抽象基类，它建立了一种参数管理模式，即每一个类的参数定义为其在构造函数__init__中的关键字参数，下面我们构造一个A类，其有两个参数x, y，并令其继承BaseClass。

```python
# 创建BaseClass的子类A
class A(BaseClass):
    
    def __init__(self, x: int = 1, y: int = 2)-> None:
        super().__init__(x=x, y=y)
```

```python
# 实例化A类，得到实例a
a = A(x=10, y=12)
```

```python
# 获取a的参数
a.get_params()
```

```python
# 修改a的参数
a.set_params(x=20, y=30)
```

```python
# 查看a修改后的参数
a.get_params()
```
