# 分组回测

回测实例参数：\
1.rate: float, 交易费用，按比例。\
2.period: int or str, 隔多少日进行调仓或者采用月初月末调仓。\
3.interest: str, 回测时使用单利还是复利, simple表示单利，compound表示复利。\
4.contract: str, 回测时使用什么合约进行交易，main表示主力合约，目前仅支持主力合约。\
5.price: str, 调仓价格, open表示开盘价（根据当天的数据计算因子，下一个交易日开盘交易），close表示收盘价。\
6.rebalance_num: int, 连续合约多少日换仓，目前支持1，3，5。\
7.start: str, 回测起始日期。\
8.end: str, 回测结束日期。\
9.group_num: int, 分组组数，默认为5。

注: 如果蚕蛹定期调仓的回测方式就会采用N日平滑。

```python
# 初始化回测实例
self = GroupBacktesting(rate=0.0003, 
                        period=20, 
                        interest='simple', 
                        contract='main', 
                        price='close', 
                        rebalance_num=1, 
                        start='2017-01-04',
                        end='2020-01-04')

# 设置因子, self.set_factor(group=因子组，name=因子名，参数1=参数1值，参数2=参数2值, ...)
self.set_factor(group='CarryFactor',name='MainNearFactor3')
# 设置商品池，self.set_commodity_pool(group=商品池组, name=商品池名, 参数1=参数1值，参数2=参数2值, ...)
self.set_commodity_pool(group='DynamicPool', name='DynamicPool3')
# 设置信号，self.set_signal(group=信号组，name=信号名, 参数1=参数1值, 参数2=参数2值, ...)
self.set_signal(group='GroupSignal', name='GroupLongSignal1', rank='ascending') # ascending表示因子值小做多
# 设置权重, self.set_weight(group=权重组, name=权重名，参数1=参数1值，参数2=参数2值, ...)
self.set_weight(group='GroupWeight', name='GroupEqualWeight')
# 运行回测
self.run_backtesting()
```

```python
self.output_backtest_result()
```

参数:\
1.period: int, default None, 表示隔多少个交易日采样。如果为None，则与GroupBacktesting实例的period保持一致。\
2.shift: int, default 0, 表示从第几个交易日开始采样。如果shift = 0, period=20, 则取第1，21，41，...个交易日。\
3.figsize: tuple, default (30, 8),表示画图的大小。\
4.heatmap_rotation1: bool, default False，第一张热力图是否旋转。\
5.heatmap_rotation2: bool, default False，第二张热力图是否旋转。\
6.annot_fontsize1: int, default 30，第一张热力图中数字字体大小。\
7.annot_fontsize2: int, default 30，第二张热力图中数字字体大小。
