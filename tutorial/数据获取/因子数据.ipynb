# 实例化因子数据管理器

因子数据位于/data/commodity_pool中，并按照类别分别保存。第一层是类别，第二层是各具体的因子的文件夹，每个具体的因子的文件夹下又有对应每种参数下的因子值DataFrame。

```python
self = FactorDataManager()
```

# 获取因子 get_factor

1.
get_factor与compute_factor的区别：\
get_factor会自动将因子数据保存到本地，而compute_factor则不会将因子数据保存到本地。

2.get_factor获取数据的优先级\
get_factor获取数据的优先级取决于参数overwrite，即是否覆盖之前的数据。\
（1）若overwrite=True，则直接根据因子的计算逻辑计算商品池，把因子实例添加到factor_dict属性中，并将因子数据保存本地。\
（2）若overwrite=False, 则先尝试从factor_dict属性中获取因子数据，再尝试从本地获取因子数据，最后尝试直接根据因子的计算逻辑计算因子，若本地无因子数据则将数据保存到本地，若factor_dict中无因子数据则将因子实例添加到factor_dict。

```python
# 获取因子的结果是一个因子实例
factor = self.get_factor(group='CarryFactor', name='MainNearFactor3')
```

# 计算因子 compute_factor

1.
get_factor与compute_factor的区别：\
get_factor会自动将因子数据保存到本地，而compute_factor则不会将因子数据保存到本地。

2.compute_factor获取数据的优先级\
compute_factor获取数据的优先级取决于参数overwrite，即是否覆盖之前的数据。\
（1）若overwrite=True，则直接根据因子的计算逻辑计算因子值，把因子实例添加到factor_dict属性中。\
（2）若overwrite=False, 则先尝试从factor_dict属性中获取因子数据，再尝试直接根据因子的计算逻辑计算因子值，若factor_dict中无因子数据则将因子实例添加到factor_dict。

```python
factor = self.compute_factor(group='CarryFactor', name='MainNearFactor3')
```

# 保存因子 save_factor

save_factor也含有参数overwrite，默认情况下overwrite=False。\
（1）如果overwrite=True, 则直接根据因子计算逻辑计算因子，将计算得到的因子实例添加到factor_dict中，并将因子实例保存到本地。\
（2）如果overwrite=False, 则先判断本地是否已经保存有因子实例，如果已经保存，则直接结束函数；如果没有保存，则先尝试从factor_dict中获取因子实例，若有，则将该因子实例的数据保存到本地，若无，则根据因子计算逻辑计算因子数据，将因子数据保存到本地。

注: 因为get_factor自带保存因子的功能，使用时用get_factorl更方便。

```python
self.save_factor(group='CarryFactor', name='MainNearFactor3')
```
