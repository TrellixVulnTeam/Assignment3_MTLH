# 实例化商品池数据管理器

商品池数据位于/data/commodity_pool中，并按照类别分别保存。第一层是类别，目前有DynamicPool和FixedPool两大类，DynamicPool分若干DynamicPool商品池的文件夹，FixedPool分若干FixedPool商品池的文件夹，每一个文件夹中会有各参数下该商品池DataFrame的pkl文件。

```python
self = CommodityPoolManager()
```

# 获取商品池 get_commodity_pool

1.
get_commodity_pool与compute_commodity_pool的区别：\
get_commodity_pool会自动将商品池数据保存到本地，而compute_commodity_pool则不会将商品池数据保存到本地。

2.get_commodity_pool获取数据的优先级\
get_commodity_pool获取数据的优先级取决于参数overwrite，即是否覆盖之前的数据。\
（1）若overwrite=True，则直接根据商品池的计算逻辑计算商品池，把商品池实例添加到commodity_pool_dict属性中，并将商品池数据保存本地。\
（2）若overwrite=False, 则先尝试从commodity_pool_dict属性中获取商品池数据，再尝试从本地获取商品池数据，最后尝试直接根据商品池的计算逻辑计算商品池，若本地无商品池数据则将数据保存到本地，若commodity_pool_dict中无商品池数据则将商品池实例添加到commodity_pool_dict。

```python
# 获取商品池的结果是一个商品池实例
dynamic_pool3 = self.get_commodity_pool(group='DynamicPool', name='DynamicPool3', q=0.3)
```

```python
# 可以看到商品池的信息，包括组别、名字和参数
dynamic_pool3
```

```python
# 通过get_params方法获取商品池的参数
dynamic_pool3.get_params()
```

```python
# 通过get_commodity_pool_value方法获取商品池数据
dynamic_pool3.get_commodity_pool_value()
```

```python
# 通过get_commodity_pool得到的商品池实例都保存在commodity_pool_dict
self.commodity_pool_dict
```

1.
get_commodity_pool与compute_commodity_pool的区别：\
get_commodity_pool会自动将商品池数据保存到本地，而compute_commodity_pool则不会将商品池数据保存到本地。

2.compute_commodity_pool获取数据的优先级\
compute_commodity_pool获取数据的优先级取决于参数overwrite，即是否覆盖之前的数据。\
（1）若overwrite=True，则直接根据商品池的计算逻辑计算商品池，把商品池实例添加到commodity_pool_dict属性中。\
（2）若overwrite=False, 则先尝试从commodity_pool_dict属性中获取商品池数据，再尝试直接根据商品池的计算逻辑计算商品池，若commodity_pool_dict中无商品池数据则将商品池实例添加到commodity_pool_dict。

```python
# 计算商品池的结果是一个商品池实例
dynamic_pool3 = self.compute_commodity_pool(group='DynamicPool', name='DynamicPool3')
```

# 保存商品池 save_commodity_pool

save_commodity_pool也含有参数overwrite，默认情况下overwrite=False。\
（1）如果overwrite=True, 则直接根据商品池计算逻辑计算商品池，将计算得到的商品池实例添加到commodity_pool_dict中，并将商品池实例保存到本地。\
（2）如果overwrite=False, 则先判断本地是否已经保存有商品池实例，如果已经保存，则直接结束函数；如果没有保存，则先尝试从commodity_pool_dict中获取商品池实例，若有，则将该商品池实例的数据保存到本地，若无，则根据商品池计算逻辑计算商品池数据，将商品池数据保存到本地。

注: 因为get_commodity_pool自带保存商品池的功能，使用时用get_commodity_pool更方便。

```python
self.save_commodity_pool(group='FixedPool', name='FixedPool1')
```

# 获取商品池中商品进出情况 get_commodity_pool_in_out

得到的结果为一个字典，包括4个键值对:\
（1）daily_in： 每日进入商品池的品种代码\
（2）daily_out: 每日退出商品池的品种代码\
（3）symbol_in_num: 每个品种在历史上进入商品池的次数\
（4）symbol_out_num: 每个品种在历史上退出商品池的次数

```python
# 每日退出商品池的品种代码
result_in_out['daily_out']
```

```python
# 品种历史上进入商品池的次数
result_in_out['symbol_in_num']
```

```python
# 品种历史上退出商品池的次数
result_in_out['symbol_out_num']
```
