# 获取全部连续合约数据

字段:\
1.datetime: datetime, 交易日期\
2.contract_before_shift: str, shift前的主力合约\
3.contract: str, 主力合约\
4.flag: float, 1.0表示换仓的第一天，0.0表示不换仓\
5.old_contract: str, 旧合约，不换仓时等于contract\
6.new_contract: str, 新合约，不换仓时等于contract\
7.old_weight: float, 旧合约的权重，不换仓时为0\
8.new_weight: float, 新合约的权重，不换仓时为1\
9.old_return: float, 旧合约的当日收益率\
10.new_return: float, 新合约的当日收益率\
11.return: float, 连续合约的当日收益率\
12.cum_return: float, 连续合约的累积收益率\
13.continuous_price: float, 连续合约的价格\
14.underlying_symbol: str, 品种代码

参数:\
1.contract: str, 基于什么合约得到的连续合约，目前仅有主力合约，即main\
2.price: str, 基于什么价格，目前支持price, close, settlement\
3.rebalance_num: int, 几日换仓，目前支持1，3，5

```python
continuous_data = self.get_continuous_contract_data(contract='main', price='close', rebalance_num=1)
```
