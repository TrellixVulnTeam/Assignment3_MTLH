# 获取所有合约基础信息

字段:\
1.contract: str, 合约代码\
2.underlying_symbol: str, 品种代码\
3.market_tplus: float, 所有期货品种都是0\
4.symbol: str, 名称+到期编号\
5.margin_rate: float, 保证金比例\
6.type: str, 类型，期货都是Future\
7.trading_code: str, 交易代码, 除郑商所外代码小写，部分品种的到期编号为三位数\
8.exhcange: str, 交易所，DCE: 大商所, CZCE: 郑商所, SHFE: 上期所, CFFEX: 中金所, INE：能源交易所\
9.product: str, 期货类型, Commodity: 商品期货, Index: 股指期货, Government: 国债期货\
10.contract_multiplier: float, 合约乘数\
11.round_lot: float, 期货全部为1.0\
12.trading_hours: str, 交易时间\
13.listed_date: datetime, 上市日期\
14.industry_name: str, 合约所属行业\
15.de_listed_date: datetime, 到期日\
16.underlying_order_book_id: 均为null
